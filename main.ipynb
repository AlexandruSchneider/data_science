```python
# imports
import os
import numpy as np
import pandas as pd
#import sklearn
#import matplotlib_inline
```

```python
# Constants
HOUSING_PATH = os.path.join("datasets", "fitness")
```

```python
# File laden
def load_data(housing_path=None) -> pd.DataFrame:
    # if no path is given, use default path to fitness directory
    if housing_path is None:
        housing_path = HOUSING_PATH
    # set the file data.csv as the file to read 
    dataPath = os.path.join(housing_path, "data.csv")
    with open(dataPath) as file:
        # return a pandas dataframe
        return pd.read_csv(file, delimiter=",")
```

```python
# Daten in Training und Test Datensatz
def split_train_test(data, test_ratio):
    # randomize the indices
    shuffled_indices = np.random.permutation(len(data))
    # get the test size based on the given test ratio
    test_set_size = int(len(data) * test_ratio)
    # generate the both lengths based on the calculated size
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
def show_scatter_matrix(train_data):
    columns = ["date", "step_count", "mood", "calories_burned", "hours_of_sleep", "bool_of_active", "weight_kg"]
    pd.plotting.scatter_matrix(train_data[columns], figsize=(13,9), diagonal="hist")
```

```python
# ----------------------------------
# Main
# ----------------------------------
# load data from data.csv file into a pandas dataframe
data_frame = load_data()
# split data into train_data(77 data rows) and test_data(22 data rows)
train_data, test_data = split_train_test(data_frame, 0.2)
show_scatter_matrix(train_data)
```
